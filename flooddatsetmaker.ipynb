```python
import os
import re
from collections import defaultdict

def sort_by_place_then_id(folder_path, suffix):
    grouped = defaultdict(list)

    files = [f for f in os.listdir(folder_path)]

    for f in files:
        lower_f = f.lower()
        
        if 'mask' in lower_f:
            # Try matching: SEN_<place>__<id>_mask.png
            match = re.match(r'SEN_([A-Za-z0-9\-_]+)__?(\d{4,5})', f, re.IGNORECASE)
            if match:
                place = match.group(1)
                uid = int(match.group(2))
                group_key = place.strip('_')
            else:
                # Try: SEN_<place>_<date>_<id>_mask.png
                match = re.match(r'SEN_([A-Za-z0-9\-_]+)_{1,2}(\d{7,8})_(\d{4})', f, re.IGNORECASE)
                if match:
                    place = match.group(1)
                    uid = int(match.group(3))
                    group_key = place.strip('_')
                else:
                    grouped["zzz"].append((99999, f))
                    continue
        elif 'before' in lower_f:
            # Handle formats for before
            match = (
                re.match(r'^SEN_([A-Za-z0-9\-_]+)_(\d{7,8})_(\d{4})_before\.(png|tif)$', f, re.IGNORECASE) or
                re.match(r'^SEN_([A-Za-z0-9\-_]+)__(\d{7,8})_(\d{4})_before\.(png|tif)$', f, re.IGNORECASE)
            )
            if match:
                place = match.group(1).strip('_')
                uid = int(match.group(3))
                date = match.group(2)
                group_key = f"{place}_{date}"
                group_key = group_key.strip('_')
                

            else:
                grouped["zzz"].append((99999, f))
                continue

        elif 'after' in lower_f:
            f = f.strip()  # Remove any leading/trailing whitespaces
        
            match = (
                re.match(r'^SEN_([A-Za-z0-9\-_]+)__(\d{4})_after\.(png|tif)$', f, re.IGNORECASE) or          # SEN_<place>__<id>_after.png
                re.match(r'^SEN_([A-Za-z0-9_\-]+)_(\d{7,8})_(\d{4})\.(png|tif)$', f, re.IGNORECASE) or       # SEN_<place>_<date>_<id>.png
                re.match(r'^SEN_([A-Za-z0-9\-_]+)_(\d{7,8})_(\d{4})_after\.(png|tif)$', f, re.IGNORECASE) or # SEN_<place>_<date>_<id>_after.png
                re.match(r'^SEN_([A-Za-z0-9\-_]+)__(\d{7,8})_(\d{4})_after\.(png|tif)$', f, re.IGNORECASE)   # SEN_<place>__<date>_<id>_after.png
            )
        
            if match:
                place = match.group(1).strip('_')
                uid = int(match.group(3)) if len(match.groups()) >= 3 else int(match.group(2))
                date = match.group(2)
                group_key = f"{place}_{date}"
                group_key = group_key.strip('_')
          
            else:
                grouped["zzz"].append((99999, f))
                continue
        elif re.match(r'^SEN_([A-Za-z0-9\-_]+)_(\d{7,8})_(\d{4})\.(png|tif)$', f, re.IGNORECASE):
            match = re.match(r'^SEN_([A-Za-z0-9\-_]+)_(\d{7,8})_(\d{4})\.(png|tif)$', f, re.IGNORECASE)
            place = match.group(1).strip('_')
            uid = int(match.group(3))
            group_key = place
        else:
            grouped["zzz"].append((99999, f))
            continue

        grouped[group_key].append((uid, f))

    final_list = []
    for place in sorted(grouped.keys()):
        sorted_group = sorted(grouped[place], key=lambda x: x[0])
        final_list.extend([fname for _, fname in sorted_group])

    print(f"[{suffix}] Total sorted files: {len(final_list)}")
    return final_list

# Folder paths
mask_path = '/kaggle/working/mask'
before_path = '/kaggle/working/before'
after_path = '/kaggle/working/after'

# Sort and save
sorted_masks = sort_by_place_then_id(mask_path, 'mask')
sorted_befores = sort_by_place_then_id(before_path, 'before')
sorted_after = sort_by_place_then_id(after_path, 'after')

with open('/kaggle/working/sorted_mask_files.txt', 'w') as f:
    f.write('\n'.join(sorted_masks))

with open('/kaggle/working/sorted_before_files.txt', 'w') as f:
    f.write('\n'.join(sorted_befores))

with open('/kaggle/working/sorted_after_files.txt', 'w') as f:
    f.write('\n'.join(sorted_after))

print("‚úÖ Sorted and saved.")
```

```python
import os
import pandas as pd
from PIL import Image
import numpy as np
import tifffile
from tqdm import tqdm

# Paths
excel_path = "/kaggle/input/selected-images/paired_data.xlsx"
before_dir = "/kaggle/working/before"
after_dir = "/kaggle/working/after"
mask_dir = "/kaggle/working/mask"

# Output dirs
input_dir = "/kaggle/working/dataset/input"
target_dir = "/kaggle/working/dataset/target"
os.makedirs(input_dir, exist_ok=True)
os.makedirs(target_dir, exist_ok=True)

# Load Excel
df = pd.read_excel(excel_path)

def load_image(path, size=(256, 256)):
    ext = os.path.splitext(path)[1].lower()
    try:
        if ext in [".tif", ".tiff"]:
            arr = tifffile.imread(path)

            if arr.dtype == np.uint16:
                arr = (arr / 256).astype(np.uint8)

            if arr.ndim == 2:
                img = Image.fromarray(arr).convert("L")
            elif arr.ndim == 3:
                img = Image.fromarray(arr).convert("L")  # Keep single channel
            else:
                raise ValueError("Unsupported TIFF shape")
        else:
            img = Image.open(path).convert("L")  # Force single channel
        return img.resize(size)
    except Exception as e:
        print(f"‚ùå Error loading {path}: {e}")
        raise

# Process images
for idx, row in tqdm(df.iterrows(), total=len(df)):
    before_path = os.path.join(before_dir, os.path.basename(row["Before"]))
    after_path = os.path.join(after_dir, os.path.basename(row["After"]))
    mask_path = os.path.join(mask_dir, os.path.basename(row["Mask"]))

    try:
        before = load_image(before_path)
        after = load_image(after_path)
        mask = load_image(mask_path)

        before_np = np.array(before) / 255.0
        after_np = np.array(after) / 255.0
        mask_np = np.array(mask) / 255.0

        # Stack before & after as channels ‚Üí shape (256, 256, 2)
        stacked_input = np.stack([before_np, after_np], axis=-1)  # (256, 256, 2)
        stacked_mask = np.expand_dims(mask_np, axis=-1)           # (256, 256, 1)

        # Save as .npy
        np.save(os.path.join(input_dir, f"input_{idx}.npy"), stacked_input)
        np.save(os.path.join(target_dir, f"target_{idx}.npy"), stacked_mask)

    except Exception as e:
        print(f"‚ö†Ô∏è Failed at index {idx}: {e}")

print("‚úÖ Dataset prepared: inputs (256x256x2), masks (256x256x1)")
```

```python
import os
from IPython.display import FileLink

def zip_and_download(directory_path, zip_filename="dataset.zip"):
    """Zips a directory and creates a downloadable link."""

    try:
        os.system(f"zip -r {zip_filename} {directory_path}")
        return FileLink(zip_filename)

    except Exception as e:
        print(f"Error zipping and downloading: {e}")
        return None

# Example usage:
directory_to_zip = "/kaggle/working/dataset"
download_link = zip_and_download(directory_to_zip)

if download_link:
    display(download_link) #display the download link.
```
